```python
import pandas as pd
import re
import pprint as pp
import string
import pickle
from wordfreq import zipf_frequency
```

```python
"""
Read the MORPHOLEX_FR file and isolate the suffixes tab
"""
fr_data = pd.read_excel(r"C:\Users\Konstantin\Documents\OB1-1\Texts\Morpholex_FR.xlsx", sheet_name="suffixes")
fr_data
```

```python
"""
Isolate the suffix column and clean it (get rid of unnecessary strings)
Then save it as a pickled file in data, with two forms, one with wordfreq that was used for German suffixes and one for summed_freq that is shown in MORHPOLEX
"""

suff_col = fr_data.iloc[:, 0]
fr_suff_dict = {key: None for key in suff_col}

# Pair the keys with summed_freq
fr_suff_dict = dict(zip(suff_col, fr_data.iloc[:, 3]))

# Iterate over the keys to clean them
str_del = ['[VB]', '\[\]', '>', '<', '{', '}', '/', '\\']
translator = str.maketrans('', '', ''.join(str_del))

for key in fr_suff_dict.copy().keys():
    new_key = key.translate(translator)
    if new_key:
        fr_suff_dict[new_key] = fr_suff_dict.pop(key)
    else:
        fr_suff_dict.pop(key, None)

fr_suff_list = list(fr_suff_dict.keys())

for key in fr_suff_list:
    fr_suff_dict[key] = zipf_frequency(key, 'fr')


with open(r'C:\Users\Konstantin\Documents\OB1-1\Data\suffix_zipf_frequency_fr.dat', 'wb') as f:
    pickle.dump(fr_suff_dict, f)

pp.pprint(fr_suff_dict)
```
