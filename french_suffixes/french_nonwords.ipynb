```python
with open(r"C:\Users\Konstantin\Documents\OB1-1\scraped_words\NONWORDS_non_stem_plus_non_suffix_french.txt", "r") as f:
    NONWORDS_no_stem_no_suffix = f.read()

with open(r"C:\Users\Konstantin\Documents\OB1-1\scraped_words\NONWORDS_non_stem_plus_suffix_french.txt", "r") as f:
    NONWORDS_no_stem_suffix = f.read()

with open(r"C:\Users\Konstantin\Documents\OB1-1\scraped_words\NONWORDS_stem_plus_non_suffix_french.txt", "r") as f:
    NONWORDS_stem_no_suffix = f.read()

with open(r"C:\Users\Konstantin\Documents\OB1-1\scraped_words\NONWORDS_stem_plus_suffix_french.txt", "r") as f:
    NONWORDS_stem_suffix = f.read()

with open(r"C:\Users\Konstantin\Documents\OB1-1\scraped_words\WORDS_morph_complex_french.txt", "r") as f:
    REALWORD_complex = f.read()

with open(r"C:\Users\Konstantin\Documents\OB1-1\scraped_words\WORDS_morph_simple_french.txt", "r") as f:
    REALWORD_simple = f.read()
```

```python
"""
Putting the scraped words into a correct format for experiments
"""

NONWORDS_no_stem_no_suffix_list = NONWORDS_no_stem_no_suffix.split()
NONWORDS_no_stem_suffix_list = NONWORDS_no_stem_suffix.split()
NONWORDS_stem_no_suffix_list = NONWORDS_stem_no_suffix.split()
NONWORDS_stem_suffix_list = NONWORDS_stem_suffix.split()
REALWORD_complex_list = REALWORD_complex.split()
REALWORD_simple_list = REALWORD_simple.split()

df1 = pd.DataFrame({'prime':NONWORDS_no_stem_no_suffix_list,'all':NONWORDS_no_stem_no_suffix_list,'target':NONWORDS_no_stem_no_suffix_list, 'condition':'non-stem/non-suffix'})
df2 = pd.DataFrame({'prime':NONWORDS_no_stem_suffix_list,'all':NONWORDS_no_stem_suffix_list,'target':NONWORDS_no_stem_suffix_list, 'condition':'non-stem/suffix'})
df3 = pd.DataFrame({'prime':NONWORDS_stem_no_suffix_list,'all':NONWORDS_stem_no_suffix_list,'target':NONWORDS_stem_no_suffix_list, 'condition':'stem/non-suffix'})
df4 = pd.DataFrame({'prime':NONWORDS_stem_suffix_list,'all':NONWORDS_stem_suffix_list,'target':NONWORDS_stem_suffix_list, 'condition':'stem/suffix'})
df5 = pd.DataFrame({'prime':REALWORD_simple_list,'all':REALWORD_simple_list,'target':REALWORD_simple_list, 'condition':'REALWORD/simple'})
df6 = pd.DataFrame({'prime':REALWORD_complex_list,'all':REALWORD_complex_list,'target':REALWORD_complex_list, 'condition':'REALWORD/complex'})


fr_df = pd.concat([df1, df2, df3, df4, df5, df6], ignore_index=True)
fr_df.insert(0, 'item_nr', range(1, len(fr_df)+1))

fr_df
```
